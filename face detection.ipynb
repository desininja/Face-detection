```python
#pass the name of picture if this code is in the same file, or pass the image location.

image = cv2.imread("IMG_5468.JPG")
```

```python
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)


#BGR2RGB is used for multi channel in hog

#BGR2GRAY is used for single channel in hog

```

FACE **DETECTION**

```python
plt.imshow(image)
```

```python
face_locations = face_recognition.face_locations(image)


number_of_faces = len(face_locations)

print("Found {} face(s) in input image.".format(number_of_faces))
```

```python
face_locations
```

```python
plt.imshow(image)
ax = plt.gca()


for face in face_locations:
    top, right, bottom, left = face
    x,y,w,h = left,top, right, bottom
    print("A face is located at pixel location")
    print("Top:",x, 'Left:',y, "Bottom:",w, "Right:",h)


    rect = Rectangle((x,y), w-x, h-y, fill = False, color = 'red')
    ax.add_patch(rect)



plt.show()
```
